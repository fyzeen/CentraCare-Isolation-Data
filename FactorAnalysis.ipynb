```python
# importing modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import FactorAnalysis
from sklearn.decomposition import PCA
from sklearn.decomposition import FastICA
```

```python
df = pd.read_csv("./data/CentracareIsolation_CLEANED.csv", index_col=0)
```

```python
data = df.copy()
for column in ["Diabetes", "Depression", "Alcoholism", "CHF", "COPD", "HTN"]:
    binarized_column = data[column].apply(lambda x: 1 if isinstance(x, str) else 0)
    data[column] = binarized_column

for column in ["Tobacco Use", "vape_user"]:
    binarized_column = data[column].apply(lambda x: 1 if (x=="Yes" or x=="Y") else 0)
    data[column] = binarized_column
```

```python
numeric = data[["Diabetes", "Depression", "Alcoholism", "CHF", "COPD", "HTN", "Tobacco Use", "vape_user", "Isolation_YN"]]
```

```python
twofactors = PCA(n_components=2)
numeric_factors = twofactors.fit_transform(numeric)
```

```python
array = np.array(data["SocialIntegrationScore"], dtype=np.int8)
colors = np.array(["Blue", "Green", "Yellow", "Red"])
```

```python
def rand_jitter(arr):
    stdev = .01 * (arr.max() - arr.min())
    return arr + np.random.randn(len(arr)) * stdev
```

```python
plt.scatter(numeric_factors[:,0], numeric_factors[:,1], c=array, alpha=0.3)
```

```python
plt.scatter(rand_jitter(numeric_factors[:,0]), rand_jitter(numeric_factors[:,1]), c=array, alpha=0.3)
```
